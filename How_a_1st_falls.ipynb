A set of statistics that analyze a player's impact on the court, hidden behind the player's choices. For instance, assisting one teammate into an open basket instead of taking a 3pt shot contested would yield positives on the advanced stats. Let's examine Jimmy Butler's advanced stats.

Again we see that Jimmy Butler also has a consistent BPM, with a dip in playoffs but mainly due to early exits, or not making it at all.BPM stands for Box Plus/Minus and it is a statistic used in basketball to estimate a player's contribution to his team's performance, relative to the league average. The league average is 0, a positive BPM indicates a player is performing above most NBA players, and a negative indicates a below-average player. 

Jimmy Butler also has an interesting ts%, with a dip in playoffs but mainly due to early exits, not making it at all, or being under more pressure to score in the playoffs. True Shooting Percentage (TS%) is a statistic in the NBA that measures a player's shooting efficiency by taking into account all types of field goal attempts (including three-point shots, two-point shots, and free throws). 

### Hypothesis Testing

We want to find out how much of an impact in efficiency Jimmy Butler had on his team and how it is opposed to other teams. It is not just to find how well Jimmy played in order to benefit the team but how efficiently he played. His totals seem to be consistent with his career in the regular season, but in the playoffs, they become more sporadic.

We decided to create a null hypothesis in which Jimmy Butler is playing much better in the playoffs in recent years than he does in the regular season. This is because, in all the data that we have shown above, we see that Jimmy Butler has had large spikes recently in his career.

Using the PPG, eFG%, TS%, and BPM, we will determine how much better Jimmy Butler is in the playoffs through standard deviations of the mean of his regular season performance.

```python
#Determine the means
rs_ppg_mean = jb_stats["PTS"].mean()
rs_efg_mean = jb_stats["eFG%"].mean()
rs_ts_mean = jb_radv["TS%"].mean()
rs_bpm_mean = jb_radv["BPM"].mean()
#Determine the standard deviations
rs_ppg_std = jb_stats["PTS"].std()
rs_efg_std = jb_stats["eFG%"].std()
rs_ts_std = jb_radv["TS%"].std()
rs_bpm_std = jb_radv["BPM"].std()
#Use playoff data
ps_ppg = jb_playoff["PTS"]
ps_efg = jb_playoff["eFG%"]
ps_ts = jb_padv["TS%"]
ps_bpm = jb_padv["BPM"]

print("Points Per Game: \n" + str((ps_ppg - rs_ppg_mean)/rs_ppg_std))

print("Efficient Field Goal %: \n" + str((ps_efg - rs_efg_mean)/rs_efg_std))

print("True Shooting Percentage: \n" + str((ps_ts - rs_ts_mean)/rs_ts_std))

print("Box Plus/Minus: \n" + str((ps_bpm - rs_bpm_mean)/rs_bpm_std))
```

### Thoughts

Based on some of our calculations, Jimmy has performed better in recent years. His standard deviations are mostly positive in recent years, so long as he makes the playoffs and has a run into the second round. With spot number 10 being the latest season, the farthest deviation into the positive is his eFG% which goes far from the mean at 3.130008 in the 2023 season. Our null hypothesis holds.

### ML/Visualizations

Ultimately we want to see how efficient Jimmy Butler truly is. By doing this, we want to determine his PER based on the factors that we discussed previously. Essentially PER (Player Efficiency Rating) is a measure of a player's overall performance. It takes into account a player's positive contributions, such as scoring, assists, and rebounds, while also accounting for negative contributions, such as missed shots and turnovers. As seen in [this](https://bleacherreport.com/articles/759714-nba-thoughts-which-stat-is-most-important-in-the-nba) article, PER is one of the most important stats to consider in an NBA player to determine their worth and effects on the court.

To determine how good Jimmy is, we took data on all the players in the current playoffs from Basketball Reference and spliced the regular and advanced stats to use as a dataset. Using a Linear Regression model and help from this sklearn information [site](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html), we want to determine how PPG, eFG%, TS% and BPM effect PER.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
#Takes in the data
nba_data = pd.read_csv("./Data/2023_nba_ml_data.csv")

# Split the dataset into training and testing sets
train_data, test_data = train_test_split(nba_data, test_size=0.2)

# Define the features and target variables
features = ['PTS', 'eFG%','TS%','BPM']
target = 'PER'

# Train the model
model = LinearRegression()
model.fit(train_data[features], train_data[target])

test_predictions = model.predict(test_data[features])

# Calculate the accuracy of the model
accuracy = r2_score(test_data[target], test_predictions)

print('The accuracy of the model is:', accuracy)
```

We see that this is fairly accurate, with most runs of this code being between the high 80s to the high 90s.

Now we will plot to see who is performing well this postseason.

```python
test_data_sorted = test_data.iloc[test_predictions.argsort()[::-1]]

# Select the top 15 players
top_15 = test_data_sorted.nlargest(15, 'PER')

colors = ['y', 'g', 'b', 'r', 'm', 'c', 'k', 'orange', 'purple', 'pink', 'brown', 'gray', 'olive', 'teal', 'navy']

# Plot the PER scores of the top 15 players
plt.bar(top_15['Player'], top_15['PER'], color=colors)
plt.xticks(rotation=90)
plt.ylabel('PER')
plt.title('Top 15 NBA Players by Predicted PER')
plt.show()
```

The plot seen above, with Jimmy Butler in the red shows that he is predicted to have one of the highest PER's at **fourth** in this dataset. This is in line with the current PER data in where teams are competing in where he is **third** currently in PER of active players. Overall, the model turned out to be fairly accurate.

Note: These findings are during the second round of the postseason as the 2023 season has not concluded, please keep this in mind as you look at this data past the 2023 season. 

### Insight

From our standard deviation, we found that Jimmy Butler has been playing at a pace that is beyond what he is playing before in the playoffs. This is while having the fourth-highest PER of postseason teams based on our findings.

Therefore, we have proven that a large factor in the Bucks losing the first seed is due to Jimmy Butler performing at his peak in recent seasons and being a league leader in efficiency among players in the postseason.

```python
# data needed to create integer values of time rounded to nearest whole number

data = [11,38,42]
  
# creates dataframe with data and creates a column called MP
df = pd.DataFrame(data, columns=['MP'])
    
# makes a bar chart with x as time and y as points 
fig, ax = plt.subplots(figsize=(8, 6))

# sets labels accordingly

ax.bar(df['MP'], giannis_playoff['PTS'])
ax.set_xlabel('Time in Minutes Rounded')
ax.set_ylabel('Total Points in Game')
ax.set_title('Comparison of Time Versus Total Points Scored')
plt.show()
                  
```

In the bar graph above we can tell that the more time he spent playing, the more he scored. This logically makes sense, but now we have data above to support this claim. He only played in three games, so there are only three bars in the graph above. We can see that in the first game he only played for around 11 minutes and did not score even half of what he did in the later two games. 

```python
# first analyze Giannis Data Alone 
fig, ax = plt.subplots(figsize=(8, 8)) 

# Create a scatter plot for MP vs Win and Loss
ax.scatter(df['MP'], giannis_playoff['WinLoss'], s=20)

ax.set_title('Time Played by Giannis vs Win or Lose By Team')
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)
plt.xlabel('Time')
plt.ylabel('Points')

# Create a line of best fit 
x = np.arange(df['MP'].min(), df['MP'].max())
m, b = np.polyfit(df['MP'], giannis_playoff['WinLoss'], 1)

# PLot that line 
ax.plot(x, m*x+b)

#show the plot 
plt.show()
```

In the graph above, we are able to get more of a confirmation that when he is playing, he is impacting the team in a big way. While we only have three games available to graph, we can clearly see an increasing slope. Meaning, the more time he spent playing, the closer the gap was if they lost. The more he played, the lower the point differential was at the end of the game if they were losing, and higher the point differential if they were winning. We only have one match showing a win. However, if we can predict what would happen at more time, we would state that the gap would be larger.
